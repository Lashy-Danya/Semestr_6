```python
import numpy as np
import matplotlib.pylab as plt
from sklearn.linear_model import LinearRegression

x = np.array([113, 124, 124, 122, 128, 140, 117, 113, 122, 139, 126, 120, 125, 118, 122]).reshape((-1, 1))
y = np.array([39, 37, 36, 36, 26, 43, 31, 40, 48, 64, 39, 34, 39, 37, 35])

model = LinearRegression()
model.fit(x, y)

y_pred = model.predict(x)

# Вывод коэффициентов модели
print(f'Коэффициент наклона (a): {model.coef_[0]}')
print(f'Смещение (b): {model.intercept_}')

# Вывод предсказанных значений
print('Предсказанные значения:', y_pred)

# Получение коэффициентов
a = model.coef_[0]
b = model.intercept_

# Вывод уравнения
equation = f'y = {a:.2f} * x + {b:.2f}'
print('Уравнение линейной регрессии:', equation)

# Графическое представление
plt.scatter(x, y, color='blue', label='Исходные данные')
plt.plot(x, y_pred, color='red', label='Линейная регрессия')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
